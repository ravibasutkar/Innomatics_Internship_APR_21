# Features of Normal Distribution 
## 1. Mean    (Central Tendency)  -   $ \mu $


## 2. Standard Deviation  (Spread) - $ \sigma $

$$ \mu = \sum_{i=1}^{N}\frac{ X_{i}}{N}$$

$$ \sigma =\sum_{i=1}^{N}\frac{ ( X_{i}-\mu)^{2}}{N}$$

# Positively Skewed & Negatively Skewed Normal Distribution

### Skewness- tells whether distribution is symmetric distribution along center vertical line or not

## If the tail of the normal distribution curve or histogram is long tail towards right => it is right (Positively) skewed distribution
## If the tail of the normal distribution curve or histogram is long tail towards left => it is left (Negatively) skewed distribution

# Explain Box Cox and show the implementation of the same

### Box Cox Transformation is to convert the non normal distributed feature to normal distribution, as Normality is an important assumption for many statistical techniques; if your data isn't normal, applying a Box-Cox means that you are able to run a broader number of tests.

```python
x_axis=dataset.Annual_HH_Income
from scipy import stats
fig = plt.figure()
ax1 = fig.add_subplot(211)

prob = stats.probplot(x_axis, dist=stats.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')

########################################################
ax2 = fig.add_subplot(212)
xt, _ = stats.boxcox(x_axis)
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
```
